```python
data['job_title'].describe()
```

```python
# Calculate average salary for each job title
average_salary_by_title = data.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)

# Select top N job titles
top_n_titles = 20
top_titles = average_salary_by_title.head(top_n_titles).index

# Filter data for top job titles
data_top_titles = data[data['job_title'].isin(top_titles)]

plt.figure(figsize=(14, 8))
sns.barplot(data=data_top_titles, x='salary_in_usd', y='job_title', errorbar=None, palette='viridis')

plt.title(f'Top {top_n_titles} Job Titles with Highest Average Salary')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')
plt.show()
```
